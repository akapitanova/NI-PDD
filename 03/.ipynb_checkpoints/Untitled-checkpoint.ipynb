```python
from sklearn.metrics import mean_squared_log_error

def rmsle(pred, test):
    """Return the RMSLe. Crop the negative predictions."""
    return np.sqrt(mean_squared_log_error(pred.clip(min=0), ytest))
```

```python
numerical = df.copy().select_dtypes(['number']).fillna(-1)
```

```python
from sklearn.model_selection import train_test_split

X = numerical.drop(['SalePrice'], axis=1, errors='ignore')
y = numerical.SalePrice

Xtrain, Xtest, ytrain, ytest = train_test_split(X, y, test_size=0.25, random_state=17)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(Xtrain, ytrain)
pred = model.predict(Xtest)
err = rmsle(pred, ytest)
```

```python
results["baseline"] = err
print(f"{results['baseline'] = }")
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_categories(df, mappers, dummies=False):
    label_encoder = LabelEncoder()
    for col in df.select_dtypes('object').columns:
        if col not in mappers and df[col].nunique() < 30:
            df[col] = df[col].fillna('NaN')
            df[col] = label_encoder.fit_transform(df[col])
            if dummies:
                df = pd.concat([df.drop(columns=[col]), pd.get_dummies(df[col], prefix=('d_' + col))], axis=1)
        elif col in mappers:
            df[col] = df[col].replace(mappers[col])
    return df
```

```python
from sklearn.decomposition import PCA
from tqdm import trange

def pca_rmsle(num_components):
    pca = PCA(n_components = num_components).fit(Xtrain)
    _Xtrain = pca.transform(Xtrain)
    _Xtest = pca.transform(Xtest)
    model = LinearRegression().fit(_Xtrain, ytrain)
    pred = model.predict(_Xtest)
    print(rmsle(pred, ytest))
    return rmsle(pred, ytest)
```

```python
res_pca = pd.DataFrame.from_records(
    ((i, pca_rmsle(i)) for i in trange(1, 170, position=0, leave=True)),
    columns=["n_components", "rmsle"]
)
```

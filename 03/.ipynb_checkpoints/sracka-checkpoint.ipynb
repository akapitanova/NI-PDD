```python
from sklearn.preprocessing import LabelEncoder

def encode_categories(df, mappers, dummies=False):
    label_encoder = LabelEncoder()
    for col in df.select_dtypes('object').columns:
        if col not in mappers and df[col].nunique() < 30:
            df[col] = df[col].fillna('NaN')
            df[col] = label_encoder.fit_transform(df[col])
            if dummies:
                df = pd.concat([df.drop(columns=[col]), pd.get_dummies(df[col], prefix=('d_' + col))], axis=1)
        elif col in mappers:
            df[col] = df[col].replace(mappers[col])
    return df
```

```python
def make_prediction(X_train, y_train, X_test, y_test):
    reg = linear_model.LinearRegression().fit(X_train, y_train)

    y_pred = reg.predict(X_test)
    rmsle = np.absolute(math.sqrt(mean_squared_log_error(y_test, y_pred)))
    #rmsle = math.sqrt(mean_squared_log_error(y_test, y_pred))
    return y_pred, rmsle
```

```python
def plot_results(y_test, y_pred):
    sns.set(font_scale=1.5)
    plt.figure(figsize=(10,6))
    sns.scatterplot(y_test, y_pred)
    plt.plot([0,600000], [0,600000], 'r')
    plt.xlabel("Prices")
    plt.ylabel("Predicted prices")
    plt.title("Prices vs. Predicted Prices")
```

```python
y_pred, rmsle = make_prediction(X_train_val, y_train_val, X_test, y_test)

print(rmsle)
```

# PCA

```python
def get_best_comps(X_train, X_val):
    best_rmsle = 50000
    best_i = 50000
    n_columns = len(X_train.columns)
    
    for i in range(1, n_columns+1):
        pca = PCA(n_components=i)
        X_train_trans = pca.fit_transform(X_train)
        #print(X_train_trans[1:4])
        X_val_trans = pca.fit_transform(X_val)
        
        try:
            y_pred, rmsle = make_prediction(X_train_trans, y_train, X_val_trans, y_val)

            if rmsle < best_rmsle:
                best_rmsle = rmsle
                best_i = i
        except:
            rmsle = math.inf
            
        print(i)
        print(rmsle)
        print()
    
    return best_rmsle, best_i
```

```python
best_rmsle, best_i = get_best_comps(X_train, X_val)
```

```python
from sklearn.feature_selection import RFE, SelectKBest, f_regression
rmsle_list = []
for n in range(1, len(X_train.columns)):
    skb = SelectKBest(f_regression, k=n)
    skb.fit(X_train, y_train)
    X_tr = skb.transform(X_train)
    X_te = skb.transform(X_test)
    reg = linear_model.LinearRegression().fit(X_tr, y_train)
    y_pred = reg.predict(X_te)
    try:
        rmsle = sqrt(mean_squared_log_error(y_test, y_pred))
    except:
        rmsle = math.inf
    rmsle_list.append(rmsle)
    print(f'n = {n}, RMSLE: {rmsle}')
```
